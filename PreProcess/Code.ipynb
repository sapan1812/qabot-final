```python
# After getting the insight from the question answers and tags dataset now create the data for usage purpose and dump 
# the data in .pkl file after creating the dataset,also options are given to choose any language from the tags dataset
# and pass them as an array in the function

# options are choosen languages, chunksize is to set the number of rows we want in the memory to load, realtve path 
# is defined by base_filepath, and paths for subsequent datasets are also passed as an argument
def PrepareTags(options, chunksize, base_filepath, tags_filepath, question_filepath, answer_filepath):
    tags_df = pd.read_csv(base_filepath + tags_filepath, sep=',') # Reading the tags dataset
    tags_df.fillna('NA', inplace=True) # Replacing the NaNs with 'NA'
    tags_df['Tag'] = tags_df['Tag'].str.lower() # Convert the Tag column to lower case
   
    for option in options:     # Looping through the options
        temp_df = tags_df[tags_df.Tag == option] 
        temp_df.sort_values(by='Id')

        # Initialized empty Dataframes
        questions_df = pd.DataFrame()
        answers_df = pd.DataFrame()
        merged_df = pd.DataFrame()

        # Checking whether there is any duplicate present and if so keeping the first one only taking 1000 or a bit
        # more number of rows
        if not temp_df['Id'].is_unique:
             temp_df.drop_duplicates("Id" , keep='first')

        # Now creating the Questions Answers dataframes, for our project, here we're only taking a little
        for chunk in pd.read_csv(base_filepath + question_filepath, chunksize=chunksize, encoding="ISO-8859-1"):
            if questions_df.shape[0] < 1000:
                temp_questions_df = pd.merge(chunk, temp_df, on='Id', how='inner')
                questions_df = questions_df.append(temp_questions_df, ignore_index=True)
            else:
                break

        for chunk in pd.read_csv(base_filepath + answer_filepath, chunksize=chunksize, encoding="ISO-8859-1"):
            if answers_df.shape[0] < 1000:
                temp_questions_df = pd.merge(chunk, temp_df, how='inner', left_on=['ParentId'], right_on=['Id'])
                answers_df = answers_df.append(temp_questions_df, ignore_index=True)
            else:
                break

        print(questions_df.columns)
        print(answers_df.columns)

        # Droping and Renaming the Columns
        questions_df.drop(['OwnerUserId', 'CreationDate', 'ClosedDate'],
                               axis=1, inplace=True)
        questions_df.rename(columns={'Body': 'Q_Body', 'Score': 'Q_Score'}, inplace=True)

        answers_df.drop(['Id_x', 'ParentId', 'CreationDate', 'OwnerUserId', 'Tag'], axis=1, inplace=True)
        answers_df.rename(columns={'Id_y': 'Id', 'Body': 'A_Body', 'Score': 'A_Score'}, inplace=True)
        
        # Merge the two dataframe to create our working dataframe
        merged_df = pd.merge(questions_df, answers_df, on='Id', how='inner')
        
        # Sort the dataframe depending on the value of Q_Score
        merged_df.sort_values('Q_Score', ascending=False, inplace=True)
        
        # Reindex dataframe
        merged_df = merged_df.reindex(['Id', 'A_Score', 'Q_Score', 'Tag', 'Title', 'Q_Body',
                                            'A_Body'], axis=1)
        
        # Names for CSV as well as Pickle files
        csv_name = option + "_latest_Questions.csv"
        pickle_name = option + "_data.pkl"
        
        # Export the dataset
        pickle.dump(merged_df, open(pickle_name, 'wb'))
        merged_df.to_csv(csv_name)
```

```python
def convert_to_lower_remove_patterns(text):
#     text = re.sub(r'@[\w]*', '', text)
    text = re.sub(r'<[^<]+?>', '', text)    # Remove the HTML tags
    text = re.sub(r'https?://\S+|www\.\S+', '', text)    # Remove the URLs tag
    for word in text.split():  # checking the contractions present and make them full form
        if word.lower() in contractions:
            text = text.replace(word, contractions[word.lower()])
    text = text.lower()  # Lower case all words
    return text
 
    
def prepare_for_communicate_patterns(text):
#     text = re.sub(r'@[\w]*', '', text)
    text = re.sub(r'<[^<]+?>', '', text) # Remove the HTML tags
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)  # Remove the URLs tag
    text = re.sub(r'[\n\r\t]', ' ', text)  # Removing the \n, \t \r
    text = re.sub(' +', ' ', text) 
    text = text.replace('&gt;', '')  # Removing >
    text = text.replace('&lt;', '')  # Removing <
    for word in text.split():  # checking the contractions present and make them full form
        if word.lower() in contractions:
            text = text.replace(word, contractions[word.lower()])
    text = text.lower() # Lower case all words
    return text

# Removal of all characters except of small/ Capital words and numbers
def remove_other_character(text):
    handle_pattern = re.compile(r"[^a-zA-Z0-9 ]")
    return handle_pattern.sub(r'', text)
    
# Remove short words (length lesser equal to 3)    
def remove_small_words(text):
    return ' '.join([w for w in text.split() if len(w) > 3])

# Removal of English STOPWORDS
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(STOPWORDS, text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

# Removal/Converion of emojis
def remove_emoji(string):
    emoji_pattern = re.compile("["
                       u"\U0001F600-\U0001F64F"  # emoticons
                       u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                       u"\U0001F680-\U0001F6FF"  # transport & map symbols
                       u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                       u"\U00002702-\U000027B0"
                       u"\U000024C2-\U0001F251"
                       "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

def convert_emojis(UNICODE_EMO, text):
    for emot in UNICODE_EMO:
        text = re.sub(r'('+emot+')', "_".join(UNICODE_EMO[emot].replace(",","").replace(":","").split()), text)
    return text

# Removal of Emoticons
emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in EMOTICONS) + u')')
def remove_emoticons(text):
    return emoticon_pattern.sub(r'', text)

# Conversion of emoticons to words
def convert_emoticons(EMOTICONS, text):
    for emot in EMOTICONS:
        text = re.sub(u'('+emot+')', "_".join(EMOTICONS[emot].replace(",","").split()), text)
    return text

# Removal of Frequent Words
def remove_frequent_words(FREQWORDS, text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

# Removal of Rare words
def remove_rare_words(RAREWORDS, text):
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

# Stemming
def stem_words(text):
    stemmer = PorterStemmer()
    return " ".join([stemmer.stem(word) for word in text.split()])

# Lemmatization
def lemmatize_words(text):
    lemmatizer = WordNetLemmatizer()
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words_list:
            new_text.append(chat_words_map_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

cnt = Counter()
```

```python
filepath = "model_answer_w2v.bin" # importing the word2vec model

from gensim.models import KeyedVectors


wv_from_bin = KeyedVectors.load_word2vec_format(filepath, binary=True) 
#extracting words vectors from our imported vector
embeddings_index = {}
for word, vector in zip(wv_from_bin.vocab, wv_from_bin.vectors):
    coefs = np.asarray(vector, dtype='float32')
    embeddings_index[word] = coefs
```

```python
def avg_feature_vector(sentence, model, num_features):
    words = sentence.split()
    #feature vector is initialized as an empty array
    feature_vec = np.zeros((num_features, ), dtype='float32')
    n_words = 0
    for word in words:
        if word in embeddings_index.keys():
            n_words += 1
            feature_vec = np.add(feature_vec, model[word])
    if (n_words > 0):
        feature_vec = np.divide(feature_vec, n_words)
    return feature_vec
```

```python
from scipy.spatial import distance
```

## Doc2Vec
### An extension of Word2Vec, the Doc2Vec embedding is one of the most popular techniques out there. Introduced in 2014, it is an unsupervised algorithm and adds on to the Word2Vec model by introducing another ‘paragraph vector’. Also, there are 2 ways to add the paragraph vector to the model.

#### 1.1) PVDM(Distributed Memory version of Paragraph Vector): We assign a paragraph vector sentence while sharing word vectors among all sentences. Then we either average or concatenate the (paragraph vector and words vector) to get the final sentence representation. It is an extension of the Continuous Bag-of-Word type of Word2Vec where we predict the next word given a set of words. It is just that in PVDM, we predict the next sentence given a set of sentences.



#### 1.2) PVDOBW( Distributed Bag of Words version of Paragraph Vector): Just lime PVDM, PVDOBW is another extension, this time of the Skip-gram type. Here, we just sample random words from the sentence and make the model predict which sentence it came from(a classification task).



```python
from gensim.models.doc2vec import LabeledSentence
from gensim.models.doc2vec import TaggedDocument
from gensim.models import Doc2Vec
from tqdm import tqdm
```

## Universal Sentence Encoder (Deep Averaging Network)
#### One of the most well-performing sentence embedding techniques right now is the Universal Sentence Encoder. And it should come as no surprise from anybody that it has been proposed by Google. The key feature here is that we can use it for Multi-task learning.

#### This means that the sentence embeddings we generate can be used for multiple tasks like sentiment analysis, text classification, sentence similarity, etc, and the results of these asks are then fed back to the model to get even better sentence vectors that before.

#### The most interesting part is that this encoder is based on two encoder models and we can use either of the two:
1) **Transformer**
2) **Deep Averaging Network(DAN)**

##### Both of these models are capable of taking a word or a sentence as input and generating embeddings for the same. The following is the basic flow:

#### Tokenize the sentences after converting them to lowercase
    1) Depending on the type of encoder, the sentence gets converted to a 512-dimensional vector
    2) If we use the transformer, it is similar to the encoder module of the transformer architecture and uses the self-attention mechanism.
    3) The DAN option computes the unigram and bigram embeddings first and then averages them to get a single embedding. This is then passed to a deep neural network to get a final sentence embedding of 512 dimensions.
    4) These sentence embeddings are then used for various unsupervised and supervised tasks like Skipthoughts, NLI, etc. The trained model is then again reused to generate a new 512 dimension sentence embedding.
    


```python
import tensorflow_hub as hub
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4" 
dan_model = hub.load(module_url)
print ("module %s loaded" % module_url)
```

```python
sentence_embeddings = dan_model(answersList)
query = user_input
query_vec = dan_model([query])[0]
```

```python
dan_position = 0
dan_score = 0
dan_marks = 0

for sent in enumerate(answersList):
    sim = cosine(query_vec, dan_model([sent[1]])[0])
    print("Sentence = ", sent[1], "; similarity = ", sim, '\n')
    
    if sim > dan_score:
        dan_score = sim
        dan_position = sent[0]
    dan_marks = dan_score * scores[sent[0]]
    
print("Marks given : ", math.ceil(dan_marks))
```
